```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Learning").getOrCreate()
```

```python
spark
```

```python
df = spark.read.csv("/content/drive/MyDrive/Pyspark/test2.csv", header=True, inferSchema=True)
df.show()
```

```python
#Removing all the null value in the dataframe
df.na.drop().show()
```

```python
#Removing the null value if any od the row is null
df.na.drop(how="any").show()
```

```python
#Removing the null value by giving thershold number
df.na.drop(how="any", thresh=2).show()
```

```python
#Removing the row of null in the particular column
df.na.drop(how="any", subset=["age"]).show()
```

```python
#Now filling technique
df.show()
```

```python
#just by replacing with missing value
df.na.fill(6).show() 
```

```python
df.na.fill("Missing_value").show()
```

Values are filling using fill method ,fills value based on dtype

```python
df.show()
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['age', 'Experience', 'Salary'],
    outputCols=["{}_imputer".format(c) for c in ["age", "Experience", "Salary"]]
).setStrategy("median")
```

```python
filled_df = imputer.fit(df).transform(df)
```

```python
filled_df.show()
```
