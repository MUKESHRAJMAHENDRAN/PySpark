```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Filtering").getOrCreate
```

```python
df = spark.read.csv("/content/drive/MyDrive/Pyspark/test1.csv", header=True, inferSchema=True)
```

```python
df.show()
```

```python
df.filter("Experience>=4").show()
```

```python
df.show()
```

```python
df.filter("age<=29").select(["Name", "Salary"]).show()
```

```python
df.filter("Experience==4" or "Experience>=10").show()
```
